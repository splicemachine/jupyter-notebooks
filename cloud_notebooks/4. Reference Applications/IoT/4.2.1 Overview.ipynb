# IoT Demo Overview
With the new Cloud Service of Splice Machine, it will be very easy to standup a powerful scale-out database that can ingest millions of records while running reports and time series analysis. We will:

* Create database for IoT in Splice Machine
* Start a Kafka stream that is a proxy for the data streams generated by Edge Bridges and other processing 
* Create a spark streaming process to retrieve data from the kafka queue
* Ingest that data into Splice Machine using the splice adapter
* Run real-time queries and reports while simultaneously continuing to ingest data into splice database
  
This demo uses our cloud infrastructure, which is based on Mesosphere DC/OS. Currently this is implemented on AWS, but it is portable to Azure, Google Cloud and can even be used for on-premise deployments. Splice plans to offer the service on these additional platforms later in 2018


## Notebooks in this Demo
There are seperate Zeppelin notebooks for this demo:

* Database Setup - to create the tables in database for IoT data. Tables are created for static and streaming data
* Kafka -  Is Kafka producer to put data on queue that will be used for the rest of the demo
* Spark Streaming - the process to capture data from Kafka and insert it into Splice Machine
* Splice Queries - running real-time queries and reports against the Splice Machine database

The notes should be run in the same sequence listed above. Each note book has details on running itself.
