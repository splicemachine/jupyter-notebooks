# IoT App Overview

Splice Machine's Cloud Service makes it easy to stand up a powerful scale-out database that can ingest millions of records while running reports and time series analysis. We have created an *Internet of Things* reference application in this section. In this app, we:

* Create a Splice Machine database for an Internet of Things project.
* Start a Kafka stream that is a proxy for the data streams generated by Edge Bridges and other processing.
* Create a spark streaming process to retrieve data from the kafka queue.
* Ingest that data into Splice Machine using the Splice Machine Native Spark DataSource.
* Run real-time queries and reports while simultaneously continuing to ingest data into our Splice Machine database.
  
This demo uses our cloud infrastructure, which is available on both AWS and Azure.
